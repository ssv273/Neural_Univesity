Как я понял, в этом алгоритме происходит следующее:  
при генерации последовательности дискриминатор предоставляет награду за выбор следующего действия, которая рассчитывается следующим образом  

$J(\theta)=\mathbb{E}\left[R_{T} \mid s_{0}, \theta\right]=\sum_{y_{1} \in \mathcal{Y}} G_{\theta}\left(y_{1} \mid s_{0}\right) \cdot Q_{D_{\phi}}^{G_{\theta}}\left(s_{0}, y_{1}\right)$

где  $Q_{D}^{G}\left(s_{0}, y_{1}\right)$ это все возможные вознаграждения, а $G\left(y_{1} \mid s_{0}\right)$ - это вероятность получить это вознаграждение.

То есть, при поступлении последовательности в дискриминатор методом Монте-Карло "развёртывается" последовательность до длины T и определяется награда, которая затем сообщается генератору.
